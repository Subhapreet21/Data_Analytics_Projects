## Indian General Elections 2024 – Voter Information Dataset

### **Dataset Overview**
This dataset provides comprehensive voter-related statistics from the **Indian General Elections 2024**, including details at the **state level**. It covers essential electoral data such as the number of electors (male, female, third gender), voter turnout percentages, postal and EVM votes, and constituency classifications (General, SC, ST). The dataset is curated from official sources and is intended to support **data analysis, visualizations, and insights into India's electoral landscape**.

### **Number of Entries**
- **36 entries** — Each representing a state or union territory.

### **Columns and Description**

| Column Name                        | Description                                                                 |
|------------------------------------|-----------------------------------------------------------------------------|
| `State Name`                       | Name of the state or union territory                                       |
| `Electors - Male`                  | Total number of male electors                                              |
| `Electors - Female`                | Total number of female electors                                            |
| `Electors - Third Gender`          | Total number of third gender electors                                      |
| `Electors - Total`                 | Combined total of all electors (male, female, third gender)                |
| `Voters - Male`                    | Number of male voters who cast their vote                                  |
| `Voters - Female`                  | Number of female voters who cast their vote                                |
| `Voters - Third Gender`            | Number of third gender voters who cast their vote                          |
| `Voters - Total`                   | Total number of actual voters (all genders)                                |
| `Voters - Poll %`                 | Percentage of voters who participated (turnout %)                          |
| `Constituency Type`                | Classification of the state’s constituencies: GEN (General), SC, ST        |
| `Postal Votes`                     | Number of postal ballots cast                                              |
| `Valid Votes Polled`              | Total number of valid votes                                                |
| `Rejected Votes (Postal)`          | Number of rejected postal ballots                                          |
| `Evm Rejected Votes`               | Number of rejected EVM votes                                               |
| `Total Rejected Votes`             | Combined total of rejected postal and EVM votes                            |

```python
state_electors = df.groupby("State Name")["Electors - Total"].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
state_electors.plot(kind='bar', color='skyblue')
plt.xticks(rotation=90)
plt.title("Total Electors by State")
plt.xlabel("State")
plt.ylabel("Number of Electors")
plt.show()
```

Generating a visualization using bar plot to display specific trends or distributions in voter data.

```python
elector_gender = df[["Electors - Male", "Electors - Female", "Electors - Third Gender"]].sum()

plt.figure(figsize=(6, 6))
plt.pie(elector_gender, labels=elector_gender.index, autopct='%1.1f%%', colors=['orange', 'yellow', 'green'])
plt.title("Gender Distribution of Electors")
plt.show()
```

Generating a visualization using Pie Chart to compare different voter attributes(gender).

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Electors - Total", y="Voters - Total", hue="State Name", alpha=0.7)
plt.title("Voter Turnout vs. Total Electors")
plt.xlabel("Total Electors")
plt.ylabel("Total Voters")
plt.legend([],[], frameon=False)  # Hide legend for better visibility
plt.show()
```

Using seaborn(scatter plot) to create a statistical visualization, revealing frequency distribution(vorter turnout vs total electors)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Voters - Poll %"], bins=15, kde=True, color="green")
plt.title("Distribution of Voter Turnout Percentage")
plt.xlabel("Voter Turnout (%)")
plt.ylabel("Frequency")
plt.show()
```

Visualize categorical data distribution(voter percentage turnout) using Seaborn(Histogram)

```python
plt.figure(figsize=(12, 6))
state_turnout = df.groupby("State Name")["Voters - Poll %"].mean().sort_values(ascending=False)

state_turnout.plot(kind='bar', color='teal')
plt.xticks(rotation=90)
plt.title("Voter Turnout Comparison by State")
plt.xlabel("State")
plt.ylabel("Voter Turnout (%)")
plt.show()
```

Generating a Matplotlib plot(bar plot), possibly representing voter turnout comaprisons

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["NOTA Votes"], bins=20, kde=True, color="red")
plt.title("Distribution of NOTA Votes")
plt.xlabel("NOTA Votes")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Total Rejected Votes"] = df["Rejected Votes (Postal)"] + df["Evm Rejected Votes"]
top_states_rejected = df.groupby("State Name")["Total Rejected Votes"].sum().nlargest(10)

plt.figure(figsize=(10, 6))
top_states_rejected.plot(kind="bar", color="red")
plt.xticks(rotation=45)
plt.title("Top 10 States with Highest Rejected Votes")
plt.xlabel("State")
plt.ylabel("Total Rejected Votes")
plt.show()
```

Generating a bar plot to find out the top 10 states with the highest rejected votes

```python
df["Postal Vote Percentage"] = (df["Voters - Postal"] / df["Voters - Total"]) * 100
postal_vote_percentage = df.groupby("State Name")["Postal Vote Percentage"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
postal_vote_percentage.plot(kind="bar", color="purple")
plt.xticks(rotation=90)
plt.title("Percentage of Postal Votes in Total Votes by State")
plt.xlabel("State")
plt.ylabel("Postal Vote Percentage (%)")
plt.show()
```

Generating a bar plot to find out the total percentage of postal votes by state

```python
plt.figure(figsize=(10, 6))
third_gender_voters = df.groupby("State Name")["Voters - Third Gender"].sum().sort_values(ascending=False)

third_gender_voters.plot(kind="bar", color="goldenrod")
plt.xticks(rotation=90)
plt.title("Third Gender Voter Participation by State")
plt.xlabel("State")
plt.ylabel("Number of Third-Gender Voters")
plt.show()
```

Generating a bar plot to find out the total number of third gender voter participation per state
