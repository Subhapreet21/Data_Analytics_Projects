## Indian General Elections 2024 – Voter Information Dataset

### **Dataset Overview**
This dataset provides comprehensive voter-related statistics from the **Indian General Elections 2024**, including details at the **state level**. It covers essential electoral data such as the number of electors (male, female, third gender), voter turnout percentages, postal and EVM votes, and constituency classifications (General, SC, ST). The dataset is curated from official sources and is intended to support **data analysis, visualizations, and insights into India's electoral landscape**.

### **Number of Entries**
- **36 entries** — Each representing a state or union territory.

### **Columns and Description**

| Column Name                        | Description                                                                 |
|------------------------------------|-----------------------------------------------------------------------------|
| `State Name`                       | Name of the state or union territory                                       |
| `Electors - Male`                  | Total number of male electors                                              |
| `Electors - Female`                | Total number of female electors                                            |
| `Electors - Third Gender`          | Total number of third gender electors                                      |
| `Electors - Total`                 | Combined total of all electors (male, female, third gender)                |
| `Voters - Male`                    | Number of male voters who cast their vote                                  |
| `Voters - Female`                  | Number of female voters who cast their vote                                |
| `Voters - Third Gender`            | Number of third gender voters who cast their vote                          |
| `Voters - Total`                   | Total number of actual voters (all genders)                                |
| `Voters - Poll %`                 | Percentage of voters who participated (turnout %)                          |
| `Constituency Type`                | Classification of the state’s constituencies: GEN (General), SC, ST        |
| `Postal Votes`                     | Number of postal ballots cast                                              |
| `Valid Votes Polled`              | Total number of valid votes                                                |
| `Rejected Votes (Postal)`          | Number of rejected postal ballots                                          |
| `Evm Rejected Votes`               | Number of rejected EVM votes                                               |
| `Total Rejected Votes`             | Combined total of rejected postal and EVM votes                            |

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="State Name", y="Voters - Poll %", palette="viridis")
plt.xticks(rotation=90)
plt.ylabel("Voter Turnout (%)")
plt.xlabel("State")
plt.title("Voter Turnout Percentage by State")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values("Electors - Total", ascending=False)
plt.bar(df_sorted["State Name"], df_sorted["Electors - Male"], label="Male", color="blue")
plt.bar(df_sorted["State Name"], df_sorted["Electors - Female"], label="Female", color="magenta", bottom=df_sorted["Electors - Male"])
plt.xticks(rotation=90)
plt.ylabel("Number of Electors")
plt.xlabel("State")
plt.legend()
plt.title("Male vs Female Electors per State")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
votes_data = [df["Valid Votes Polled"].sum(), df["Rejected Votes (Postal)"].sum() + df["Evm Rejected Votes"].sum()]
labels = ["Valid Votes", "Rejected Votes"]
plt.pie(votes_data, labels=labels, autopct="%1.1f%%", colors=["green", "red"])
plt.title("Valid vs Rejected Votes Distribution")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df_sorted = df.sort_values("Electors - Total", ascending=False)
plt.bar(df_sorted["State Name"], df_sorted["Electors - Total"], label="Total Electors", color="skyblue")
plt.bar(df_sorted["State Name"], df_sorted["Voters - Total"], label="Total Voters", color="orange")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.xlabel("State")
plt.legend()
plt.title("Total Electors vs. Total Voters per State")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Voters - Poll %"], bins=15, kde=True, color="purple")
plt.xlabel("Voter Turnout (%)")
plt.ylabel("Number of States")
plt.title("Distribution of Voter Turnout Percentage")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(data=df, x="State Name", y="NOTA Votes", palette="Reds_r")
plt.xticks(rotation=90)
plt.ylabel("NOTA Votes")
plt.xlabel("State")
plt.title("NOTA Votes per State")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Electors - Total", y="Voters - Poll %", hue="State Name", palette="Set2", s=100)
plt.xlabel("Total Electors")
plt.legend(loc="center right")
plt.ylabel("Voter Turnout (%)")
plt.title("Scatter Plot of Electors vs. Voter Turnout",loc="left")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Constituency Type", y="Voters - Poll %", palette="coolwarm")
plt.xlabel("Constituency Type")
plt.ylabel("Voter Turnout (%)")
plt.title("Voter Turnout by Constituency Type")
plt.show()
```

```python
top_states = df.nlargest(10, "Voters - Poll %")
plt.figure(figsize=(12, 6))
sns.barplot(data=top_states, x="State Name", y="Voters - Poll %", palette="Blues_r")
plt.xticks(rotation=45)
plt.ylabel("Voter Turnout (%)")
plt.xlabel("State")
plt.title("Top States by Voter Turnout Percentage")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x="State Name", y="Voters - Poll %", marker="o", color="blue")
plt.xticks(rotation=90)
plt.ylabel("Voter Turnout (%)")
plt.xlabel("State")
plt.title("Voter Turnout Percentage by State")
plt.show()
```
