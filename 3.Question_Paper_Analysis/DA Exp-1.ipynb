```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('class_marks.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df["Q1"]=df["Q1aM4"]+df["Q1bM6"] 
df["Q2"]=df["Q2aM6"]+df["Q2bM4"] 
df["Q3"]=df["Q3aM5"]+df["Q3bM5"] 
df["Q4"]=df["Q4aM3"]+df["Q4bM7"] 
df["Q5"]=df["Q5M10"] 
df["Q6"]=df["Q6aM4"]+df["Q6bM6"] 
df.drop(["Q1aM4","Q1bM6","Q2aM6","Q2bM4","Q3aM5","Q3bM5","Q4aM3","Q4bM7","Q5M10","Q6aM4","Q6bM6"],axis=1,inplace=True)
df
```

```python
df.info()
```

```python
sns.histplot(df['Total'], kde=True, bins=10, color='blue').set(title='Distribution of Total Scores', xlabel='Total Marks', ylabel='Frequency')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Q1'], y=df['Q2'])
plt.title('Scatter Plot of Q1 vs Q2')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Total'])
plt.title('Count of Each Category')
plt.xticks(rotation=45)
plt.show()
```

```python
def assign_grade(marks):
    if marks >= 50:
        return 'A'
    elif marks >= 40:
        return 'B'
    elif marks >= 30:
        return 'C'
    elif marks >= 20:
        return 'D'
    else:
        return 'F'
df['Grade'] = df['Total'].apply(assign_grade)
df.head()
```

```python
plt.figure(figsize=(6, 6))
df['Grade'].value_counts().plot.pie(autopct='%1.1f%%', cmap='coolwarm', startangle=90)
plt.ylabel('')
plt.title('Distribution of Grades')
plt.show()
```

```python
def pass_or_fail(marks):
    return 'Pass' if marks >= 18 else 'Fail'
df['pass_fail'] = df['Total'].apply(pass_or_fail)
df.head()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x=df['pass_fail'], palette=['green', 'red'])
plt.title('Pass vs Fail Count')
plt.xlabel('Result')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['Total'])
plt.title('Box Plot of Total Marks')
plt.ylabel('Marks')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Q1', 'Q2', 'Q3','Q4','Q5','Q6']])
plt.title('Box Plot of Subject-Wise Marks')
plt.ylabel('Marks')
plt.show()
```

```python
def performance_category(grade):
    if grade == 'A':
        return 'Excellent'
    elif grade == 'B':
        return 'Good'
    elif grade == 'C':
        return 'Average'
    else:
        return 'Poor'

df['Performance'] = df['Grade'].apply(performance_category)
df
```

```python
plt.figure(figsize=(6, 6))
df['Performance'].value_counts().plot.pie(autopct='%1.1f%%', cmap='cool', startangle=90)
plt.ylabel('')
plt.title('Performance Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Grade'], palette='magma', order=['A', 'B', 'C', 'D', 'F'])
plt.title('Number of Students in Each Grade')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df_grouped = df.groupby('Grade')[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']].mean()
df_grouped.T.plot(kind='bar', figsize=(10, 5), colormap='plasma')
plt.title('Average Marks Per Question by Grade')
plt.xlabel('Questions')
plt.ylabel('Average Marks')
plt.legend(title="Grade")
plt.show()
```
